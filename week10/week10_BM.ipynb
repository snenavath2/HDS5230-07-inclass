```python
import matplotlib.pyplot as plt

stages = ['Dummy', 'KNN Training', 'KNN Prediction', 'KNN Retraining', 'KNN Retrained Prediction']
accuracy = [0.5075, 0.895, 0.963, 0.721, 0.965]
recall = [0.5048, 0.8867, 0.9407, 0.4163, 0.913]
precision = [0.5327, 0.9045, 0.9061, 1.0, 0.8873]

plt.figure(figsize=(10, 6))
x = list(range(len(stages)))
plt.plot(x, accuracy, 'bo-', linewidth=2, label='Accuracy')
plt.plot(x, recall, 'gs-', linewidth=2, label='Recall')
plt.plot(x, precision, 'r^-', linewidth=2, label='Precision')

for i in range(1, len(stages)):
    plt.axvline(x=i-0.5, color='gray', linestyle='--', alpha=0.5)

plt.title('Model Performance Across Stages')
plt.xlabel('Stage')
plt.ylabel('Score')
plt.xticks(x, stages, rotation=15)
plt.ylim(0.4, 1.05)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```
